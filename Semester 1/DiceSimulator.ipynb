```python
import seaborn as sns
import matplotlib.pyplot as plt
import random

die_sides = 6

def roll_die(sides):
    return random.randint(1, sides)

die_rolls = [0, 0, 0, 0, 0, 0]
for roll in range(100):
    die_rolls[roll_die(die_sides)-1] += 1

sns.barplot(x=["1", "2", "3", "4", "5", "6"], y=die_rolls)
plt.xlabel("Die Roll")
plt.ylabel("Frequency")
plt.title("Frequency of Each Die Roll Within a Simulation of 100 Rolls")
plt.show()
```
